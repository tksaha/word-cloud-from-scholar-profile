```python
from scholarly import scholarly
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
def get_publications(author_name):
    search_query = scholarly.search_author(author_name)
    author = scholarly.fill(next(search_query))
    return author.get("publications")
```

```python
def extract_titles(publications):
    titles = []
    for pub in publications:
        titles.append(pub['bib']['title'].replace("using", " "))
    return titles
```

```python
def generate_word_cloud(titles):
    text = ' '.join(titles)
    wordcloud = WordCloud(width=800, height=200, background_color='white', max_words=50).generate(text)
    return wordcloud
```

```python
def display_word_cloud(wordcloud):
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
author_name = "Tanay Kumar Saha"  # Replace with the name of the Google Scholar profile you want to analyze
publications = get_publications(author_name)
titles = extract_titles(publications)
wordcloud = generate_word_cloud(titles)
display_word_cloud(wordcloud)
```
